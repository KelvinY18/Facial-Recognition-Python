```python
import cv2

```

```python
# specify the location of images and HaarCascades files
cascade_file = 'haarcascade_frontalface_default.xml'
image_file = 'neutral (2).jpg'

# create the HaarCascade object
face_cascade = cv2.CascadeClassifier(cascade_file)

# read the image
image = cv2.imread(image_file)

# convert the image into grayscale
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
# BGR --> Blue Green Red --> standard color format for pictures taken by a camera



# detect face in an Image
# detectMultiScale() --> to detect objects from image

# firstParameter --> grayScale image

# secondParameter --> scaleFactor
# in scaleFactor we are going to resize the image
# using scaleFactor we can resize a large face into a smaller one

# thirdParameter --> minNeighbors = 3
# this will affet the quality of the detected face

# forthParameter --> minSize = (30,30)
# this parameter determines how small size image we want to detect

faces = face_cascade.detectMultiScale(gray,scaleFactor=1.3,minNeighbors=3,minSize=(30,30))



# draw a rectangle around face
# cv2.rectangle(originalImage,startPoint,endpoint,color,thickness)
for x,y,w,h in faces:
  cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)


# display final output
from google.colab.patches import cv2_imshow
cv2_imshow(image)

```
