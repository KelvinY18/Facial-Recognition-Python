```python
import cv2
# specify the location of images and haarcascades files

cascade_file = 'haarcascade_frontalface_default.xml'
image_file = 'Nazwa.jpg'

# create the haarcascade object
face_cascade = cv2.CascadeClassifier(cascade_file)

# read the image
image = cv2.imread(image_file)

# before process any image, we need to convert it to grayscale image
gray = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
# BGR --> blue green red
# BGR is a standard color format for pictures taken by a camera

# detect face in an image
# detectMultiScale()--> use to detect objects from an image
# firstparameter --> grayscale image
# secondparameter --> scalefactor (resize the image)
# thirdparameter --> minNeighbors = 3
# this will effect the quality of the detected face
# forthparameter --> minSize
# this parameter determines how small size image that want to detect

faces = face_cascade.detectMultiScale(gray,scaleFactor=1.3,minNeighbors=3,minSize=(30,30))

# draw a rectangle around face
# cv2.rectangle(originalimage , starpoint , endpoint , color , thickness)

for x,y,w,h in faces:
  cv2.rectangle(image, (x,y) , (x+w , y+h) , (0,255,0) , 2)

# display the final output

from google.colab.patches import cv2_imshow
cv2_imshow(image)
```
