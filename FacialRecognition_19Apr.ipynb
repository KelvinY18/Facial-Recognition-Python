```python
import cv2
from google.colab.patches import cv2_imshow

# specify the location of images and haarCascade file

cascadeFile = 'haarcascade_frontalface_default.xml'
imageFile = 'caprice.png'

# create the haarCascade object

faceCascade = cv2.CascadeClassifier(cascadeFile)

# read the image

image = cv2.imread(imageFile)

# convert image into grayscale

gray = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
# BGR --> Blue Green Red
# BGR is a standard color format for pictures taken by a camera

# detect face in an image
# detectMultiScale() --> used to detect objects
# firstParameter --> grayscale image
# secondParameter --> scaleFactor
# in scaleFactor we are going to resize the image
# using scaleFactor you can resize a larger face into a smaller one
# making it detectable by the algorithm
# thirdParameter --> minNeighbors
# this parameter will affect the quality of the detected faces
# forthParameter --> minSize
# this parameter determines how small size image you want to detect

faces = faceCascade.detectMultiScale(gray , scaleFactor=1.5 , minNeighbors=3 , minSize=(30,30))

# draw a rectangle around faces
# cv2.rectangle(originalImage , startPoint , endPoint , color , thickness)

for x,y,w,h in faces:
  cv2.rectangle(image , (x,y) , (x+w , y+h) , (0,255,0) , 2)

# display the final output
cv2_imshow(image)
```
